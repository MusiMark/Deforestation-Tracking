```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", 120)
```

```python
df = pd.read_csv("../data/raw/deforestation_tracking_dataset.csv")

df.head()
```

```python
missing_summary = pd.DataFrame({
    "Missing Values": df.isna().sum(),
    "Missing Percentage (%)": (df.isna().mean() * 100).round(2),
})

missing_summary = missing_summary.sort_values("Missing Percentage (%)", ascending=False)
missing_summary
```

## Missing Data Treatment Decision

The dataset has only a tiny amount of missing data: 1 value each in `Annual_Rainfall_mm`, `Illegal_Logging`, and `Forest_Area_ha` (0.07% per feature). Because the missingness is so sparse, the best default choice is **removal of the affected rows** rather than imputation.

Why this choice:
- The amount of missing data is negligible, so dropping a few rows will not meaningfully reduce the dataset.
- Imputation would add synthetic values for only three cells, which is unlikely to improve the analysis.
- Using algorithms that handle missing values natively is unnecessary here because the missingness is too small to justify changing the modeling pipeline.

```python
target_columns = ["Annual_Rainfall_mm", "Illegal_Logging", "Forest_Area_ha"]
df_cleaned = df.dropna(subset=target_columns)
```

```python
missing_summary = pd.DataFrame({
    "Missing Values": df_cleaned.isna().sum(),
})

missing_summary = missing_summary.sort_values("Missing Values", ascending=False)
missing_summary
```

```python
file_path = os.path.join("..", "data", "cleaned", "deforestation_cleaned.csv")

folder_path = os.path.dirname(file_path)

df_cleaned.to_csv(file_path, index=False)

print(f"Data successfully saved to relative path: {file_path}")
```

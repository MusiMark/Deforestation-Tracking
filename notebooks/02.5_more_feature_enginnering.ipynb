```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib

df = pd.read_csv("../data/engineered/deforestation_features.csv")

# Drop unnecessary columns
df.drop(columns=["Record_ID"], inplace=True)

# Remove duplicate features
df.drop(columns=["Illegal_Logging","Population_Pressure","Fire_Logging_Impact"], inplace=True)

# Date
df["Observation_Date"] = pd.to_datetime(df["Observation_Date"], errors="coerce")

df["Year"] = df["Observation_Date"].dt.year
df["Month"] = df["Observation_Date"].dt.month
df["Day"] = df["Observation_Date"].dt.day

df.drop(columns=["Observation_Date"], inplace=True)

# One-Hot Encoding
df = pd.get_dummies(
    df,
    columns=["Region", "District", "Forest_Type"],
    dtype=int
)

# Drop new null records
df.dropna(axis=0, inplace=True)


# Label Encoding
categorical_columns = "Deforestation_Risk"

encoder = LabelEncoder()
df[categorical_columns] = encoder.fit_transform(df[categorical_columns])

# Save all encoders
joblib.dump(encoder, "../src/label_encoder.pkl")

# Saving dataset

output_file = "../data/engineered/deforestation_training.csv"

df.to_csv(output_file, index=False)

print(f"Dataset saved successfully to {output_file}")
```
